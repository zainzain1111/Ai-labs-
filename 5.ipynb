```python
# -*- coding: utf-8 -*-
"""
LAB Assignment No. 5: Artificial Neural Network Model
This script implements solutions for all 7 questions of the lab assignment
using NumPy, Scikit-learn, Matplotlib, and TensorFlow/Keras,
suitable for execution in a Google Colab environment.
"""

import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Flatten
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.datasets import (
    load_iris,
    load_breast_cancer,
    fetch_california_housing,
)
from sklearn.metrics import accuracy_score, mean_squared_error


# Set random seed for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

# =============================================================================
# Question 1: Logic Gates (AND) with Neural Network
# =============================================================================
print("--- Question 1: AND Gate Implementation ---")

# 1. Create dataset using NumPy
X_and = np.array([[0, 0], [0, 1], [1, 0], [1, 1]], dtype=np.float32)
y_and = np.array([0, 0, 0, 1], dtype=np.float32)

print(f"AND Inputs (X):\n{X_and}")
print(f"AND Outputs (y):\n{y_and}")

# 2. Build a neural network with one hidden layer (2 neurons)
model_and = Sequential([
    # Hidden layer with 2 neurons and ReLU activation
    Dense(2, activation='relu', input_shape=(2,)),
    # Output layer with 1 neuron and Sigmoid activation for binary output
    Dense(1, activation='sigmoid')
])

# Compile the model
model_and.compile(optimizer='adam',
                  loss='binary_crossentropy',
                  metrics=['accuracy'])

# Show model structure
model_and.summary()

# 3. Train it
print("\nTraining the AND gate model...")
history_and = model_and.fit(X_and, y_and, epochs=100, verbose=0)

# Show accuracy
final_accuracy_and = history_and.history['accuracy'][-1]
print(f"Final Training Accuracy: {final_accuracy_and:.4f}")

# 4. Compare model predictions with actual outputs
predictions_and = model_and.predict(X_and)
# Convert probabilities to binary output (0 or 1)
binary_predictions_and = (predictions_and > 0.5).astype(int).flatten()

print("\n--- Model Prediction Comparison (AND Gate) ---")
print(f"{'Input':<10}{'Actual':<10}{'Predicted':<10}")
for i in range(len(X_and)):
    print(f"{str(X_and[i]):<10}{y_and[i]:<10.0f}{binary_predictions_and[i]:<10.0f}")

# =============================================================================
# Question 2: Regression Task (y = x^2 + noise)
# =============================================================================
print("\n" + "="*80)
print("--- Question 2: Regression (y = x^2 + noise) ---")

# 1. Generate 100 samples
n_samples = 100
X_reg = np.linspace(-3, 3, n_samples).reshape(-1, 1)  # X input (feature)
noise = np.random.normal(0, 0.5, size=(n_samples, 1))
y_reg = X_reg**2 + noise  # y = x^2 + noise (target)

# Split data
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(
    X_reg, y_reg, test_size=0.2, random_state=42)

# 2. Build a neural network to predict y from x (Base Model: 10 neurons)
def build_reg_model(n_neurons):
    model = Sequential([
        Dense(n_neurons, activation='relu', input_shape=(1,)),
        Dense(n_neurons, activation='relu'),
        Dense(1) # Output layer for regression (linear activation)
    ])
    model.compile(optimizer='adam', loss='mse', metrics=['mae'])
    return model

# Train the base model
model_reg = build_reg_model(n_neurons=10)
print("\nTraining Base Regression Model (10 hidden neurons)...")
history_reg = model_reg.fit(X_train_reg, y_train_reg,
                            epochs=100, verbose=0, validation_split=0.1)

# Predict
y_pred_reg = model_reg.predict(X_reg)

# 3. Plot actual vs. predicted results
plt.figure(figsize=(10, 6))
plt.scatter(X_reg, y_reg, label='Actual Data ($y=x^2 + noise$)', color='blue', alpha=0.6)
plt.plot(X_reg, y_pred_reg, label='Model Predictions', color='red', linewidth=3)
plt.title('Q2: Regression of $y = x^2 + noise$')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# 4. Discuss how increasing hidden neurons changes results
print("\n--- Discussion: Impact of Hidden Neurons on Regression ---")

neurons_list = [2, 10, 100]
results_q2 = {}

for n in neurons_list:
    model_test = build_reg_model(n_neurons=n)
    model_test.fit(X_train_reg, y_train_reg, epochs=100, verbose=0)
    y_pred_test = model_test.predict(X_test_reg)
    mse = mean_squared_error(y_test_reg, y_pred_test)
    results_q2[n] = mse
    print(f"Model with {n:3d} neurons | Test MSE: {mse:.4f}")

print("\nObservation:")
print("- With fewer neurons (e.g., 2), the model might underfit, unable to capture the parabolic curve, resulting in higher MSE.")
print("- With a moderate number (e.g., 10), the model is typically able to fit the curve well.")
print("- With excessive neurons (e.g., 100), the model risks overfitting the noise in the training data, which might not significantly improve test MSE and could generalize poorly to new data.")

# =============================================================================
# Question 3: XOR Gate with Different Activation Functions
# =============================================================================
print("\n" + "="*80)
print("--- Question 3: XOR Gate Activation Function Comparison ---")

# Dataset for XOR (The classic non-linear problem)
X_xor = np.array([[0, 0], [0, 1], [1, 0], [1, 1]], dtype=np.float32)
y_xor = np.array([0, 1, 1, 0], dtype=np.float32)

def train_and_evaluate_xor(activation_func, epochs=200):
    model = Sequential([
        Dense(4, activation=activation_func, input_shape=(2,)),
        Dense(1, activation='sigmoid') # Sigmoid for binary output
    ])

    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

    # Train and capture history
    history = model.fit(X_xor, y_xor, epochs=epochs, verbose=0)
    loss = history.history['loss'][-1]
    acc = history.history['accuracy'][-1]

    print(f"Activation: {activation_func:<8} | Final Accuracy: {acc:.4f} | Final Loss: {loss:.4f} | Epochs to Converge: {len(history.history['loss'])}")
    return history.history['loss'], activation_func

# Train with different activations
activations = ['sigmoid', 'tanh', 'relu']
histories_q3 = []

for act in activations:
    print(f"\nRunning XOR with {act} activation...")
    loss_history, name = train_and_evaluate_xor(act)
    histories_q3.append((loss_history, name))

# Plot results
plt.figure(figsize=(10, 6))
for loss_history, name in histories_q3:
    plt.plot(loss_history, label=f'{name.capitalize()} Loss')

plt.title('Q3: XOR Gate Loss Convergence by Activation Function')
plt.xlabel('Epochs')
plt.ylabel('Binary Crossentropy Loss')
plt.legend()
plt.grid(True, alpha=0.3)
plt.ylim(0, 1) # Set Y limit for clearer comparison
plt.show()

print("\nDiscussion:")
print("- Tanh often converges fastest for XOR because its output range is centered at 0 (-1 to 1), making gradient flow more efficient than sigmoid (0 to 1).")
print("- Sigmoid can suffer from the vanishing gradient problem, especially if weights are large, leading to slower convergence.")
print("- ReLU is typically faster than sigmoid but may struggle slightly with simple binary inputs, though it often performs well overall and avoids the vanishing gradient problem.")

# =============================================================================
# Question 4: Binary Classification (Breast Cancer Dataset)
# =============================================================================
print("\n" + "="*80)
print("--- Question 4: Binary Classification (Breast Cancer) ---")

# Load dataset
data_bc = load_breast_cancer(as_frame=True)
X_bc, y_bc = data_bc.data, data_bc.target

# Split data
X_train_bc, X_test_bc, y_train_bc, y_test_bc = train_test_split(
    X_bc, y_bc, test_size=0.2, random_state=42)

# Preprocessing: Scaling features is crucial for NNs
scaler_bc = MinMaxScaler()
X_train_scaled_bc = scaler_bc.fit_transform(X_train_bc)
X_test_scaled_bc = scaler_bc.transform(X_test_bc)

# Build Neural Network
model_bc = Sequential([
    Dense(32, activation='relu', input_shape=(X_train_scaled_bc.shape[1],)),
    Dense(16, activation='relu'),
    # Output layer: 1 neuron, sigmoid for binary classification
    Dense(1, activation='sigmoid')
])

# Compile
model_bc.compile(optimizer='adam',
                 loss='binary_crossentropy',
                 metrics=['accuracy'])

# Train
print("\nTraining Breast Cancer model...")
history_bc = model_bc.fit(X_train_scaled_bc, y_train_bc,
                          epochs=50, batch_size=32, verbose=0, validation_split=0.1)

# Evaluate
loss_bc, acc_bc = model_bc.evaluate(X_test_scaled_bc, y_test_bc, verbose=0)
print(f"Test Loss (Breast Cancer): {loss_bc:.4f}")
print(f"Test Accuracy (Breast Cancer): {acc_bc:.4f}")

# =============================================================================
# Question 5: Multi-Class Classification (Iris Dataset)
# =============================================================================
print("\n" + "="*80)
print("--- Question 5: Multi-Class Classification (Iris) ---")

# Load dataset
data_iris = load_iris(as_frame=True)
X_iris, y_iris = data_iris.data, data_iris.target
n_classes = len(np.unique(y_iris))

# Preprocessing: Scaling and One-Hot Encoding
scaler_iris = StandardScaler()
X_scaled_iris = scaler_iris.fit_transform(X_iris)
# Convert labels to one-hot encoding (e.g., 0 -> [1,0,0], 1 -> [0,1,0], etc.)
y_one_hot_iris = to_categorical(y_iris, num_classes=n_classes)

# Split data
X_train_iris, X_test_iris, y_train_iris, y_test_iris = train_test_split(
    X_scaled_iris, y_one_hot_iris, test_size=0.3, random_state=42)

# Build Neural Network
model_iris = Sequential([
    Dense(10, activation='relu', input_shape=(X_train_iris.shape[1],)),
    Dense(5, activation='relu'),
    # Output layer: 3 neurons (one for each class), Softmax for probability distribution
    Dense(n_classes, activation='softmax')
])

# Compile (Use categorical_crossentropy for multi-class, one-hot encoded labels)
model_iris.compile(optimizer='adam',
                   loss='categorical_crossentropy',
                   metrics=['accuracy'])

# Train
print("\nTraining Iris model...")
history_iris = model_iris.fit(X_train_iris, y_train_iris,
                              epochs=50, batch_size=8, verbose=0, validation_split=0.1)

# Evaluate
loss_iris, acc_iris = model_iris.evaluate(X_test_iris, y_test_iris, verbose=0)
print(f"Test Loss (Iris): {loss_iris:.4f}")
print(f"Test Accuracy (Iris): {acc_iris:.4f}")

# =============================================================================
# Question 6: Regression Problem (California Housing)
# =============================================================================
print("\n" + "="*80)
print("--- Question 6: Regression (California Housing) ---")

# Load dataset
data_housing = fetch_california_housing(as_frame=True)
X_house, y_house = data_housing.data, data_housing.target

# Split data
X_train_house, X_test_house, y_train_house, y_test_house = train_test_split(
    X_house, y_house, test_size=0.2, random_state=42)

# Preprocessing: Scaling features (StandardScaler recommended for regression)
scaler_house = StandardScaler()
X_train_scaled_house = scaler_house.fit_transform(X_train_house)
X_test_scaled_house = scaler_house.transform(X_test_house)

# Build Neural Network (Deeper model for complex regression)
model_house = Sequential([
    Dense(64, activation='relu', input_shape=(X_train_scaled_house.shape[1],)),
    Dense(32, activation='relu'),
    Dense(16, activation='relu'),
    # Output layer: 1 neuron, linear activation for price prediction
    Dense(1)
])

# Compile (Use MSE and MAE for regression)
model_house.compile(optimizer='adam',
                    loss='mse',
                    metrics=['mae'])

# Train
print("\nTraining California Housing model...")
history_house = model_house.fit(X_train_scaled_house, y_train_house,
                                epochs=30, batch_size=64, verbose=0, validation_split=0.1)

# Evaluate
loss_house, mae_house = model_house.evaluate(X_test_scaled_house, y_test_house, verbose=0)
print(f"Test Mean Squared Error (MSE): {loss_house:.4f}")
# MAE is easier to interpret: average prediction error in thousands of dollars
print(f"Test Mean Absolute Error (MAE) (in $100k): {mae_house:.4f}")

# =============================================================================
# Question 7: Neural Network with Dropout Regularization (MNIST)
# =============================================================================
print("\n" + "="*80)
print("--- Question 7: Neural Network with Dropout (MNIST) ---")

# Load MNIST dataset
(X_train_mnist, y_train_mnist), (X_test_mnist, y_test_mnist) = tf.keras.datasets.mnist.load_data()

# Preprocessing
# 1. Normalize pixel values to 0-1
X_train_norm = X_train_mnist.astype('float32') / 255.0
X_test_norm = X_test_mnist.astype('float32') / 255.0

# 2. Convert labels to one-hot encoding (10 classes)
y_train_one_hot = to_categorical(y_train_mnist, num_classes=10)
y_test_one_hot = to_categorical(y_test_mnist, num_classes=10)

# Build Neural Network with Dropout
model_dropout = Sequential([
    # Flatten the 28x28 image into a 784-dimensional vector
    Flatten(input_shape=(28, 28)),
    Dense(512, activation='relu'),
    # Dropout layer: randomly sets 30% of input units to 0 at each update
    Dropout(0.3),
    Dense(256, activation='relu'),
    # Another Dropout layer
    Dropout(0.3),
    Dense(10, activation='softmax') # 10 classes output
])

# Compile
model_dropout.compile(optimizer='adam',
                      loss='categorical_crossentropy',
                      metrics=['accuracy'])

# Train (Use validation_data to monitor for overfitting)
print("\nTraining MNIST model with Dropout...")
history_dropout = model_dropout.fit(X_train_norm, y_train_one_hot,
                                    epochs=20,
                                    batch_size=128,
                                    verbose=0,
                                    validation_data=(X_test_norm, y_test_one_hot))

# Plot training vs validation loss
plt.figure(figsize=(10, 6))
plt.plot(history_dropout.history['loss'], label='Training Loss')
plt.plot(history_dropout.history['val_loss'], label='Validation Loss')
plt.title('Q7: MNIST Model Loss with Dropout')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

print("\nDiscussion on Dropout Regularization:")
print("- Dropout is a regularization technique where randomly selected neurons are ignored during training (e.g., setting their weights to zero).")
print("- By introducing Dropout (e.g., 30% rate), the model is prevented from relying too heavily on specific neurons, which forces the network to learn more robust, distributed representations.")
print("- This typically results in the Validation Loss (blue line) staying closer to or slightly above the Training Loss (orange line) for longer, mitigating the gap that indicates overfitting in a standard network.")
print(f"Final Test Accuracy: {history_dropout.history['val_accuracy'][-1]:.4f}")

# =============================================================================
# END OF ASSIGNMENT
# =============================================================================
```
